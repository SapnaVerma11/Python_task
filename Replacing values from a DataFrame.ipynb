### Create a DataFrame and replace some of the values

```python
import pandas as pd
import numpy as np
students = np.random.randint(low = 2000,high = 50000,size = 10)
rooms = np.random.randint(low = 20,high = 30,size = 10 )
dict1 = {
    "Students":students,
    "Rooms":rooms
}
df = pd.DataFrame(dict1)
df
```

### Replacing a single value using replace method


```python
df.replace(24,30)
```

### Replacing multiple values

```python

df.replace({45917:22222,13083:11111})
```
