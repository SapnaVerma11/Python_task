### Create a DataFrame, save it as a Csv File, set one of the column as index 



```python
#creating a dataframe

import pandas as pd
import numpy as np

area = np.random.randint(low = 2000,high = 5000,size = 20)
price = np.random.randint(low = 200000,high = 900000,size = 20)
rooms = np.random.randint(low = 10,high = 20,size = 20)
windows = np.random.randint(low = 20,high = 40,size = 20)   
dict1 = {
    "Area":area,
    "Price":price,
    "Rooms":rooms,
    "Windows":windows
    
}
df = pd.DataFrame(dict1)
df
```

```python
#saving it to a csv file
df.to_csv("price_data")
```

```python
df.set_index('Windows',inplace = True)
df
```
